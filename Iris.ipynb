**Spliting Data into input Column and output column**

```python
data = df.values
```

```python
X = data[:,0:4]
Y = data[:,4]
```

```python
print(X)
print(Y)
```

**Testing and Train The Data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
print(X_test)
```

```python
print(Y_test)
```

**The First Training Model using Vector Machine Algorithm**

```python
from sklearn.svm import SVC
Model_svc= SVC()
Model_svc.fit(X_train,Y_train)
```

**Prediction Using Previous Model**
**And Checking The Accuracy**

```python
Predict1 = Model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,Predict1)*100)
```

**print The Output Prediction with Previous Model**

```python
for i in range (len(Predict1)):
    print(Y_test[i],Predict1[i])
```

**!!So there Are Some Wrong Prediction in Many Output**

**Second Model**

```python
#Logistic Regrestion
from sklearn.linear_model import LogisticRegression
Model_LR = LogisticRegression(solver='liblinear')
Model_LR.fit(X_train,Y_train)
```

**Accuracy Check**

```python
Predict2 = Model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,Predict2)*100)
```

**Classification Report**

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Predict2))
```
